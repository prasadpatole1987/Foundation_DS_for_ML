# What are inputs and outputs?

### Goal: Train a model to make accurate predictions

### Objective function: Defines the taks and measures how well the model performs

### Data: Inputs given to the model and correct answers needed by the objective functions.

### Optimizer: Ajusts the model's parameters based on errors so it improves over time.

### Key point: objective, data and optimizer are only needed during training; trainig updates parameters but does not change the model type.

```python
# Load data with pandas
import pandas as pd

dataset = pd.read_csv('doggy-boot-harness.csv')

dataset.head()
```

```python
# Filte data by columns

print('Harness Sizes')
print(dataset.harness_size)

# Deleting column from datasets
del dataset['sex']
del dataset['age_years']

# Print remaining columns after deleting age and sex
print('\nAvailable columns after deleting sex and age information:')
print(dataset.columns.values)
```

```python
# Display Top of tables
print(dataset.head())

# Display bottom of tables
print(dataset.tail())
```

```python
# print length of dataset
print(f"We have {len(dataset)} rows of data")

is_small = dataset.harness_size < 55
print(is_small)

data_from_small_dogs = dataset[is_small]
print(data_from_small_dogs)
print(f'Number of dogs with harness size less than 55: {len(data_from_small_dogs)}')
```

```python
# Make a copy of data that only contains dogs with a boot size below 40
data_smaller_paws = dataset[dataset.boot_size < 40].copy()

data_smaller_paws.tail()
```

```python
# Create a graph for data

import matplotlib.pyplot as plt

plt.scatter(data_smaller_paws['harness_size'], data_smaller_paws['boot_size'])

plt.xlabel('Harness Size')

plt.ylabel('Boot Size')
```
